### 1) What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?

I would look at the rate at which driver partners accessed the tolls before and after implementing the experiment

## 2) Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on: 
### a) how you will implement the experiment
I would assign an ID to each driver and keep track of how many times each driver accessed the toll booth, or went through the toll booth. If the ultimate goal is to have the drivers serving both cities, then they should be going back and forth, as such. 

### b) what statistical test(s) you will conduct to verify the significance of the observation
Since it is a Qualitative to Quantitative relationshipI would perform T-tests to determine the significance of reimbursing tolls on the number of driver partners who access the tolls. 

### c) how you would interpret the results and provide recommendations to the city operations team along with any caveats.
I would look at the p-value of the results. If the value is less than 0.05 than we can conclude with reasonable assurance that the number of times a driver accessed a toll in a day was not randomly related to the reimbursement of tolls, but directly related

# SVM

```python
from sklearn import model_selection
from sklearn import svm, metrics
from sklearn.svm import SVC
from sklearn.metrics import plot_confusion_matrix, accuracy_score
from mlxtend.plotting import plot_decision_regions
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
svm = SVC(random_state=42, kernel='rbf')

svm = svm.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

matrix = plot_confusion_matrix(svm, X_test, y_test,
                                 cmap=plt.cm.Blues,
                                 normalize='true')
plt.title('Confusion matrix for RBF SVM')
plt.show(matrix)
plt.show()

y_pred = svm.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Model accuracy: {accuracy}')

print(classification_report(y_test, y_pred))
```

```python
print("SVM")
print("Accuracy:", metrics.accuracy_score(y_test,y_pred))
print("Balanced accuracy:", metrics.balanced_accuracy_score(y_test,y_pred))
print('Precision score' , metrics.precision_score(y_test,y_pred))
print('Recall score' , metrics.recall_score(y_test,y_pred))
```
