- Now i will do some simple forecasting of the different categories (fruit, vegetable, cut_flowers, pot_plants)
- I will have to remove some features since I am using the fbprophet model
- I only need a date (ds) and y feature (price)
- Since I am not specifying for each type of thing (each type of fruit, each type of vegetable, etc...) this forecast will be very rough and only show estimates for each type of item

```python
from fbprophet import Prophet

from fbprophet.diagnostics import cross_validation, performance_metrics
from fbprophet.plot import plot_cross_validation_metric
```

```python
fruit = fruit.rename(columns={'date': 'ds', 'price': 'y'})
vegetable = vegetable.rename(columns={'date': 'ds', 'price': 'y'})
cut_flowers = cut_flowers.rename(columns={'date': 'ds', 'price': 'y'})
pot_plants = pot_plants.rename(columns={'date': 'ds', 'price': 'y'})
```

## Fruit Model

```python
fruit.shape
```

```python
fr_model = Prophet()

fr_model.fit(fruit.drop(columns=['category', 'item', 'variety', 'unit', 'month', 'year']))
```

```python
fr_future = fr_model.make_future_dataframe(periods=365)

fr_pred = fr_model.predict(fr_future)

fr_model.plot(fr_pred)
fr_model.plot_components(fr_pred)
```

```python
fr_cv = cross_validation(fr_model, initial = '1490 days', horizon = '10 days')
fr_cv.head()
```

```python
fr_p = performance_metrics(fr_cv)
fr_p.head()
```

```python
print(fr_p.rmse.mean())
print(fr_p.rmse.max() - fr_p.rmse.min())
```

```python
print('fr Model predicts with RMSE Score:', fr_p.rmse.mean())
```

## Vegetable Model

```python
ve_model = Prophet()

ve_model.fit(vegetable.drop(columns=['category', 'item', 'variety', 'unit', 'month', 'year']))
```

```python
ve_future = ve_model.make_future_dataframe(periods=365)

ve_pred = ve_model.predict(ve_future)

ve_model.plot(ve_pred)
ve_model.plot_components(ve_pred)
```

```python
ve_cv = cross_validation(ve_model, initial = '1490 days', horizon = '10 days')
ve_cv.head()
```

```python
ve_p = performance_metrics(ve_cv)
print(ve_p.head())

print(ve_p.rmse.mean())
print(ve_p.rmse.max() - ve_p.rmse.min())
```

```python
print('Veg Model predicts with RMSE Score:', ve_p.rmse.mean())
```

## cut_flowers model

```python
cf_model = Prophet()

cf_model.fit(cut_flowers.drop(columns=['category', 'item', 'variety', 'unit', 'month', 'year']))

cf_future = cf_model.make_future_dataframe(periods=365)

cf_pred = cf_model.predict(cf_future)

cf_model.plot(cf_pred)
cf_model.plot_components(cf_pred)

cf_cv = cross_validation(cf_model, initial = '1490 days', horizon = '10 days')
print(cf_cv.head())

cf_p = performance_metrics(cf_cv)
print(cf_p.head())

print(cf_p.rmse.mean())
print(cf_p.rmse.max() - cf_p.rmse.min())

```

```python
print('cf Model predicts with RMSE Score:', cf_p.rmse.mean())
```

## pot_plants model

```python
pp_model = Prophet()

pp_model.fit(pot_plants.drop(columns=['category', 'item', 'variety', 'unit', 'month', 'year']))

pp_future = pp_model.make_future_dataframe(periods=365)

pp_pred = pp_model.predict(pp_future)

pp_model.plot(pp_pred)
pp_model.plot_components(pp_pred)

pp_cv = cross_validation(pp_model, initial = '45 days', horizon = '10 days')
print(pp_cv.head())

pp_p = performance_metrics(pp_cv)
print(pp_p.head())

print(pp_p.rmse.mean())
print(pp_p.rmse.max() - pp_p.rmse.min())
```
